# Data Cleaning!

#### Drop NaN columns

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

#### Droping the rows where Amount ==NaN

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"]=df["Amount"].astype('int')
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df[['Age','Amount']].describe()
```

# Exploratory Data Analysis!

### Quetions: Which gender did more Shopping?

```python
gf1=sns.countplot(x='Gender',data=df)
for bar in gf1.containers:
    gf1.bar_label(bar)
    
```

```python
sale_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sale_gender
```

```python
sns.barplot(x='Gender',y='Amount',data=sale_gender)
```

##### from the above graphs we can seem that most of the buyers are females even the females are spend more than men.

# Age

```python
gf2=sns.countplot(x='Age Group',data=df,hue='Gender')
for bar in gf2.containers:
    gf2.bar_label(bar)
```

```python
sale_age=df.groupby(['Age Group'],as_index=False )['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
gf3=sns.barplot(x='Age Group',y='Amount',data=sale_age)
```

#### From the above graph most of the buyers are age group between 26-35 years females

```python
Sales_state_order=df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
```

```python
gf4=sns.barplot(x="State",y="Orders",data=Sales_state_order)
sns.set(rc={"figure.figsize":(15,10)})
gf4.set_xticklabels(gf4.get_xticklabels(), rotation=45)
gf4
```

```python
Sales_state=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
gf5=sns.barplot(x='State',y='Amount',data=Sales_state)
gf5.set_xticklabels(gf4.get_xticklabels(),rotation=45)
gf5
```

#### From above grahp we can see that most of the orders are from "Uttar Pradesh", "Maharashtra" and "Karnataka" respectively.

```python
gf6=sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(6,4)})
for bar in gf6.containers:
    gf6.bar_label(bar)
```

```python
sales_married=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_married
```

```python
gf7=sns.barplot(x='Marital_Status',y='Amount',data=sales_married,hue='Gender')
```

#### From above graph we can see that most the buyers are married (Women) and they have high purchasing power.

# Occupation

```python
gf8=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,10)})
```

```python
sales_occu=df.groupby(["Occupation"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
gf9=sns.barplot(x='Occupation',y='Amount',data=sales_occu)
```

#### From the above graph we can see the most the buyers are working in it, Healthcare, and Aviation sectors.

```python
df.columns
```

```python
gf10=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,10)})
gf10.set_xticklabels(gf10.get_xticklabels(),rotation=45)

for bar in gf10.containers:
    gf10.bar_label(bar)
```

```python
sales_Pro=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
gf11=sns.barplot(x='Product_Category',y='Amount',data=sales_Pro)
```

#### From above graph most sold products are Food, cloths & Apparel, and Electronics

```python
sales_pro_Id=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
gf12=sns.barplot(x="Product_ID",y='Orders',data=sales_pro_Id)
gf12.set_xticklabels(gf12.get_xticklabels(),rotation=45)
gf12
```
